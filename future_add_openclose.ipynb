```python
import pandas as pd
from enum import Enum, auto

class open_interest_delta_forward_enum(Enum):
    NONE = auto()
    UNKNOWN = auto()
    EXCHANGE = auto()
    OPENFWDOUBLE = auto()
    OPEN = auto()
    CLOSEFWDOUBLE = auto()
    CLOSE = auto()

class order_forward_enum(Enum):
    UP = auto()
    DOWN = auto()
    MIDDLE = auto()

class tick_type_key_enum(Enum):
    TICKTYPE = auto()
    TICKCOLOR = auto()

class tick_type_enum(Enum):
    NOCHANGE = '无交易'
    UNKNOWN = '未知'
    EXCHANGELONG = '多换'
    EXCHANGESHORT = '空换'
    EXCHANGEUNKOWN = '未知换'
    OPENDOUBLE = '双开'
    OPENLONG = '多开'
    OPENSHORT = '空开'
    OPENUNKOWN = '未知开'
    CLOSEDOUBLE = '双平'
    CLOSESHORT = '空平'
    CLOSELONG = '多平'
    CLOSEUNKOWN = '未知平'

class tick_color_enum(Enum):
    WHITE = '白色'
    RED = '红色'
    GREEN = '绿色'

tick_type_cal_dict = {
    open_interest_delta_forward_enum.NONE:
        {
            order_forward_enum.UP: {tick_type_key_enum.TICKTYPE: tick_type_enum.NOCHANGE,
                                    tick_type_key_enum.TICKCOLOR: tick_color_enum.WHITE},
            order_forward_enum.DOWN: {tick_type_key_enum.TICKTYPE: tick_type_enum.NOCHANGE,
                                      tick_type_key_enum.TICKCOLOR: tick_color_enum.WHITE},
            order_forward_enum.MIDDLE: {tick_type_key_enum.TICKTYPE: tick_type_enum.NOCHANGE,
                                        tick_type_key_enum.TICKCOLOR: tick_color_enum.WHITE}
        },
    open_interest_delta_forward_enum.EXCHANGE:
        {
            order_forward_enum.UP: {tick_type_key_enum.TICKTYPE: tick_type_enum.EXCHANGELONG,
                                    tick_type_key_enum.TICKCOLOR: tick_color_enum.RED},
            order_forward_enum.DOWN: {tick_type_key_enum.TICKTYPE: tick_type_enum.EXCHANGESHORT,
                                      tick_type_key_enum.TICKCOLOR: tick_color_enum.GREEN},
            order_forward_enum.MIDDLE: {tick_type_key_enum.TICKTYPE: tick_type_enum.EXCHANGEUNKOWN,
                                        tick_type_key_enum.TICKCOLOR: tick_color_enum.WHITE}
        },
    open_interest_delta_forward_enum.OPENFWDOUBLE:
        {
            order_forward_enum.UP: {tick_type_key_enum.TICKTYPE: tick_type_enum.OPENDOUBLE,
                                    tick_type_key_enum.TICKCOLOR: tick_color_enum.RED},
            order_forward_enum.DOWN: {tick_type_key_enum.TICKTYPE: tick_type_enum.OPENDOUBLE,
                                      tick_type_key_enum.TICKCOLOR: tick_color_enum.GREEN},
            order_forward_enum.MIDDLE: {tick_type_key_enum.TICKTYPE: tick_type_enum.OPENDOUBLE,
                                        tick_type_key_enum.TICKCOLOR: tick_color_enum.WHITE}
        },
    open_interest_delta_forward_enum.OPEN:
        {
            order_forward_enum.UP: {tick_type_key_enum.TICKTYPE: tick_type_enum.OPENLONG,
                                    tick_type_key_enum.TICKCOLOR: tick_color_enum.RED},
            order_forward_enum.DOWN: {tick_type_key_enum.TICKTYPE: tick_type_enum.OPENSHORT,
                                      tick_type_key_enum.TICKCOLOR: tick_color_enum.GREEN},
            order_forward_enum.MIDDLE: {tick_type_key_enum.TICKTYPE: tick_type_enum.OPENUNKOWN,
                                        tick_type_key_enum.TICKCOLOR: tick_color_enum.WHITE}
        },
    open_interest_delta_forward_enum.CLOSEFWDOUBLE:
        {
            order_forward_enum.UP: {tick_type_key_enum.TICKTYPE: tick_type_enum.CLOSEDOUBLE,
                                    tick_type_key_enum.TICKCOLOR: tick_color_enum.RED},
            order_forward_enum.DOWN: {tick_type_key_enum.TICKTYPE: tick_type_enum.CLOSEDOUBLE,
                                      tick_type_key_enum.TICKCOLOR: tick_color_enum.GREEN},
            order_forward_enum.MIDDLE: {tick_type_key_enum.TICKTYPE: tick_type_enum.CLOSEDOUBLE,
                                        tick_type_key_enum.TICKCOLOR: tick_color_enum.WHITE}
        },
    open_interest_delta_forward_enum.CLOSE:
        {
            order_forward_enum.UP: {tick_type_key_enum.TICKTYPE: tick_type_enum.CLOSESHORT,
                                    tick_type_key_enum.TICKCOLOR: tick_color_enum.RED},
            order_forward_enum.DOWN: {tick_type_key_enum.TICKTYPE: tick_type_enum.CLOSELONG,
                                      tick_type_key_enum.TICKCOLOR: tick_color_enum.GREEN},
            order_forward_enum.MIDDLE: {tick_type_key_enum.TICKTYPE: tick_type_enum.CLOSEUNKOWN,
                                        tick_type_key_enum.TICKCOLOR: tick_color_enum.WHITE}
        },
}

 # Mapping of openclose and price_direction to open_interest_delta_forward_enum and order_forward_enum
openclose_mapping = {
    '换手': open_interest_delta_forward_enum.EXCHANGE,
    '双开': open_interest_delta_forward_enum.OPENFWDOUBLE,
    '开仓': open_interest_delta_forward_enum.OPEN,
    '双平': open_interest_delta_forward_enum.CLOSEFWDOUBLE,  # Assuming definition for 双平
    '平仓': open_interest_delta_forward_enum.CLOSE,
    '无交易': open_interest_delta_forward_enum.NONE,
    '未知':open_interest_delta_forward_enum.UNKNOWN
}

price_direction_mapping = {
    '向上': order_forward_enum.UP,
    '向下': order_forward_enum.DOWN,
    '不变': order_forward_enum.MIDDLE
}

# 枚举到中文的映射
tick_type_to_chinese = {
    tick_type_enum.NOCHANGE: '无交易',
    tick_type_enum.UNKNOWN: '未知',
    tick_type_enum.EXCHANGELONG: '多换',
    tick_type_enum.EXCHANGESHORT: '空换',
    tick_type_enum.EXCHANGEUNKOWN: '未知换',
    tick_type_enum.OPENDOUBLE: '双开',
    tick_type_enum.OPENLONG: '开多',
    tick_type_enum.OPENSHORT: '开空',
    tick_type_enum.OPENUNKOWN: '未知开',
    tick_type_enum.CLOSEDOUBLE: '双平',
    tick_type_enum.CLOSESHORT: '空平',
    tick_type_enum.CLOSELONG: '多平',
    tick_type_enum.CLOSEUNKOWN: '未知平',
}

tick_color_to_chinese = {
    tick_color_enum.WHITE: '白色',
    tick_color_enum.RED: '红色',
    tick_color_enum.GREEN: '绿色',
}


# Load the tick data
df = pd.read_csv("future_taobao_ss2404_tick.csv")

# Calculate the deltas for volume and open interest
df['volume_delta'] = df['volume'].diff()
df['open_interest_delta'] = df['open_interest'].diff()

# Initialize columns for open/close position
df['trade_openclose'] = None
df['trade_color'] = None 
df['close_count'] = df['volume_delta']

# Iterate over the DataFrame to assign open/close positions
for i in range(1, len(df)):
    volume_delta = df.loc[i, 'volume_delta']
    open_interest_delta = df.loc[i, 'open_interest_delta']
    last_price = df.loc[i, 'last_price']
    ask_price1 = df.loc[i, 'ask_price1']
    bid_price1 = df.loc[i, 'bid_price1']
    pre_last_price = df.loc[i-1,'last_price']
    pre_ask_price1 = df.loc[i-1, 'ask_price1']
    pre_bid_price1 = df.loc[i-1, 'bid_price1']
    openclose = ''
    price_direction = ''

    # Determine the open/close position
    if volume_delta == open_interest_delta > 0:
        openclose = '双开'
    elif volume_delta > open_interest_delta > 0:
        openclose = '开仓'
    elif volume_delta > abs(open_interest_delta) > 0 and open_interest_delta < 0:
        openclose = '平仓'
    elif volume_delta > 0 and open_interest_delta == 0:
        openclose = '换手'
    elif volume_delta + open_interest_delta == 0 and volume_delta>0:
        openclose = '双平'
    elif volume_delta == 0 and  open_interest_delta == 0:
        openclose = '无交易'
    else:
        openclose = '未知'
    
    # 判断价格方向
    if last_price >= pre_ask_price1:
        price_direction = '向上'
    elif last_price <= pre_bid_price1:
        price_direction = '向下'
    elif last_price >= ask_price1:
        price_direction = '向上'
    elif last_price <= bid_price1:
        price_direction = '向下'
    else:
        price_direction = '不变'
        
   

    # Convert openclose and price_direction to their corresponding enum values
    delta_enum = openclose_mapping.get(openclose, open_interest_delta_forward_enum.NONE)
    direction_enum = price_direction_mapping.get(price_direction, order_forward_enum.MIDDLE)

    # Retrieve the tick type and color based on the mappings
    tick_info = tick_type_cal_dict.get(delta_enum, {}).get(direction_enum, {})
    tick_type = tick_info.get(tick_type_key_enum.TICKTYPE, tick_type_enum.UNKNOWN)
    tick_color = tick_info.get(tick_type_key_enum.TICKCOLOR, tick_color_enum.WHITE)
    

    # 使用映射转换为中文描述
    tick_type_chinese = tick_type_to_chinese[tick_type]
    tick_color_chinese = tick_color_to_chinese[tick_color]

    # 更新DataFrame
    df.at[i, 'trade_openclose'] = tick_type_chinese
    df.at[i, 'trade_color'] = tick_color_chinese

# Save the modified DataFrame to a new CSV file
df.to_csv("future_taobao_ss2404_tick_withopenclose.csv", index=False)
```

```python
import pandas as pd

# 假设已加载并准备好DataFrame df
df = pd.read_csv("future_taobao_ss2404_tick_withopenclose.csv")

df = df.loc[df['volume_delta'] > 0]
# 重置索引以确保从0开始
df.reset_index(drop=True, inplace=True)

# 初始化列用于标记交易机会
df['buy_opportunity'] = 0
df['sell_opportunity'] = 0
trade_hand=1

# 为每个tick计算未来100个tick内与当前挂单价格相匹配的成交量
for i in range(len(df)):
    if i + 100 >= len(df):  # 确保未来有足够的tick进行分析
        break

    # 当前的买一和卖一价
    current_ask_price1 = df.loc[i, 'ask_price1']
    current_bid_price1 = df.loc[i, 'bid_price1']

    # 初始化对应价格的成交量计数
    matching_buy_volume = 0
    matching_sell_volume = 0
    
    buy_done = 0
    sell_done = 0

    # 遍历未来100个tick
    for j in range(i+1, i+100):
        if df.loc[j, 'bid_price1']<current_bid_price1:
            buy_done = 1 
        if df.loc[j, 'ask_price1']>current_ask_price1:
            sell_done = 1 
        # 如果未来tick的买一价等于当前tick的买一价，累加成交量
        if df.loc[j, 'bid_price1'] == current_bid_price1 and df.loc[j, 'bid_price1'] == df.loc[j, 'last_price']:
            matching_buy_volume += df.loc[j, 'volume_delta']
        # 如果未来tick的卖一价等于当前tick的卖一价，累加成交量
        if df.loc[j, 'ask_price1'] == current_ask_price1 and df.loc[j, 'ask_price1'] == df.loc[j, 'last_price']:
            matching_sell_volume += df.loc[j, 'volume_delta']

    # 判断是否存在买卖机会
    if buy_done==1 or matching_buy_volume >= df.loc[i, 'ask_volume1']+trade_hand:
        df.loc[i, 'buy_opportunity'] = 1
    if sell_done==1 or matching_sell_volume >= df.loc[i, 'bid_volume1']+trade_hand:
        df.loc[i, 'sell_opportunity'] = 1


# 创建一个新列来标记同时为买卖机会的tick
df['dual_opportunity'] = ((df['buy_opportunity'] == 1) & (df['sell_opportunity'] == 1)).astype(int)

# 计算同时为买卖机会的总数
dual_opportunities_count = df['dual_opportunity'].sum()

# 计算比例
total_ticks_analyzed = len(df) - 100
dual_opportunities_ratio = dual_opportunities_count / total_ticks_analyzed

print(f"Dual opportunities count: {dual_opportunities_count}")
print(f"Dual opportunities ratio: {dual_opportunities_ratio:.4f}")

# 保存到新的CSV文件
df.to_csv("future_taobao_ss2404_tick_with_opportunities.csv", index=False)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score

# 加载数据
data = pd.read_csv("future_taobao_ss2404_tick_with_opportunities.csv")

data_clean = data.dropna().copy()  # 创建一个副本以避免警告

features = [
    'last_price','highest','lowest', 'volume', 'open_interest', 'volume_delta', 'open_interest_delta',
    'bid_price1', 'ask_price1', 'bid_price2', 'ask_price2', 'bid_price3', 'ask_price3',
    'bid_price4', 'ask_price4', 'bid_price5', 'ask_price5', 'bid_volume1', 'bid_volume2',
    'bid_volume3', 'bid_volume4', 'bid_volume5', 'ask_volume1', 'ask_volume2', 'ask_volume3',
    'ask_volume4', 'ask_volume5'
]
# 计算短期量价特征
window_size = 100
data_clean['price_sma'] = data_clean['last_price'].rolling(window=window_size).mean()
data_clean['volume_sma'] = data_clean['volume'].rolling(window=window_size).mean()
data_clean['price_change_rate'] = data_clean['last_price'].pct_change(periods=window_size)
data_clean['volume_change_rate'] = data_clean['volume'].pct_change(periods=window_size)
data_clean['price_std'] = data_clean['last_price'].rolling(window=window_size).std()
data_clean['highest_price'] = data_clean['highest'].rolling(window=window_size).max()
data_clean['lowest_price'] = data_clean['lowest'].rolling(window=window_size).min()
data_clean['price_volume_corr'] = data_clean[['last_price', 'volume']].rolling(window=window_size).corr().iloc[0::2, -1].reset_index(level=1, drop=True)

data_clean = data_clean.dropna()

# 更新特征列表
features.extend([
    'price_sma', 'volume_sma', 'price_change_rate', 'volume_change_rate',
    'price_std', 'highest_price', 'lowest_price', 'price_volume_corr'
])

# 目标列处理
data_clean['successful_trade'] = data_clean.apply(
    lambda row: 1 if row['buy_opportunity'] and row['sell_opportunity'] else 0, axis=1
)

X = data_clean[['datetime'] + features]
y = data_clean['successful_trade']

# 数据切分
split_point = int(len(data_clean) * 0.8)
X_train = X.iloc[:split_point]
y_train = y.iloc[:split_point]
X_test = X.iloc[split_point:]
y_test = y.iloc[split_point:]

# 模型训练
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
# 注意：随机森林模型不能直接处理datetime类型的特征，所以在训练前需要去除或转换该特征
X_train_features = X_train.drop(columns=['datetime'])
X_test_features = X_test.drop(columns=['datetime'])
rf_model.fit(X_train_features, y_train)

# 模型评估
predictions = rf_model.predict(X_test_features)
print("Model Accuracy:", accuracy_score(y_test, predictions))

# 对照组评估（全1预测）
predictions_all_ones = [1] * len(y_test)
print("All Ones Accuracy:", accuracy_score(y_test, predictions_all_ones))
```
