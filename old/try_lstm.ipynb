```python
import mplfinance as mpf
import numpy as np
import pandas as pd

# Reload the data
daily_data = pd.read_csv('daily_002594.csv')

# Convert 'trade_date' to datetime type and set it as index
daily_data['trade_date'] = pd.to_datetime(daily_data['trade_date'], format='%Y%m%d')
daily_data.set_index('trade_date', inplace=True)

# Select the columns required for mplfinance and rename them
daily_data_mpf = daily_data[['open', 'high', 'low', 'close', 'vol']]
daily_data_mpf.columns = ['Open', 'High', 'Low', 'Close', 'Volume']

# Create the K line chart
mpf.plot(daily_data_mpf[-90:], type='candle', style='yahoo', volume=True, title='K Line Chart (Last 90 Days)', figratio=(12,6))



```
