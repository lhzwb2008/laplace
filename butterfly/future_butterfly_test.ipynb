```python
#快速版本
import pandas as pd
import numpy as np
from enum import Enum, auto

FILENAME = "future_taobao_ss2405"

class open_interest_delta_forward_enum(Enum):
    NONE = auto()
    UNKNOWN = auto()
    EXCHANGE = auto()
    OPENFWDOUBLE = auto()
    OPEN = auto()
    CLOSEFWDOUBLE = auto()
    CLOSE = auto()

class order_forward_enum(Enum):
    UP = auto()
    DOWN = auto()
    MIDDLE = auto()

class tick_type_key_enum(Enum):
    TICKTYPE = auto()
    TICKCOLOR = auto()

class tick_type_enum(Enum):
    NOCHANGE = '无交易'
    UNKNOWN = '未知'
    EXCHANGELONG = '多换'
    EXCHANGESHORT = '空换'
    EXCHANGEUNKOWN = '未知换'
    OPENDOUBLE = '双开'
    OPENLONG = '开多'
    OPENSHORT = '开空'
    OPENUNKOWN = '未知开'
    CLOSEDOUBLE = '双平'
    CLOSESHORT = '空平'
    CLOSELONG = '多平'
    CLOSEUNKOWN = '未知平'

class tick_color_enum(Enum):
    WHITE = '白色'
    RED = '红色'
    GREEN = '绿色'

tick_type_cal_dict = {
    open_interest_delta_forward_enum.NONE:
        {
            order_forward_enum.UP: {tick_type_key_enum.TICKTYPE: tick_type_enum.NOCHANGE,
                                    tick_type_key_enum.TICKCOLOR: tick_color_enum.WHITE},
            order_forward_enum.DOWN: {tick_type_key_enum.TICKTYPE: tick_type_enum.NOCHANGE,
                                      tick_type_key_enum.TICKCOLOR: tick_color_enum.WHITE},
            order_forward_enum.MIDDLE: {tick_type_key_enum.TICKTYPE: tick_type_enum.NOCHANGE,
                                        tick_type_key_enum.TICKCOLOR: tick_color_enum.WHITE}
        },
    open_interest_delta_forward_enum.EXCHANGE:
        {
            order_forward_enum.UP: {tick_type_key_enum.TICKTYPE: tick_type_enum.EXCHANGELONG,
                                    tick_type_key_enum.TICKCOLOR: tick_color_enum.RED},
            order_forward_enum.DOWN: {tick_type_key_enum.TICKTYPE: tick_type_enum.EXCHANGESHORT,
                                      tick_type_key_enum.TICKCOLOR: tick_color_enum.GREEN},
            order_forward_enum.MIDDLE: {tick_type_key_enum.TICKTYPE: tick_type_enum.EXCHANGEUNKOWN,
                                        tick_type_key_enum.TICKCOLOR: tick_color_enum.WHITE}
        },
    open_interest_delta_forward_enum.OPENFWDOUBLE:
        {
            order_forward_enum.UP: {tick_type_key_enum.TICKTYPE: tick_type_enum.OPENDOUBLE,
                                    tick_type_key_enum.TICKCOLOR: tick_color_enum.RED},
            order_forward_enum.DOWN: {tick_type_key_enum.TICKTYPE: tick_type_enum.OPENDOUBLE,
                                      tick_type_key_enum.TICKCOLOR: tick_color_enum.GREEN},
            order_forward_enum.MIDDLE: {tick_type_key_enum.TICKTYPE: tick_type_enum.OPENDOUBLE,
                                        tick_type_key_enum.TICKCOLOR: tick_color_enum.WHITE}
        },
    open_interest_delta_forward_enum.OPEN:
        {
            order_forward_enum.UP: {tick_type_key_enum.TICKTYPE: tick_type_enum.OPENLONG,
                                    tick_type_key_enum.TICKCOLOR: tick_color_enum.RED},
            order_forward_enum.DOWN: {tick_type_key_enum.TICKTYPE: tick_type_enum.OPENSHORT,
                                      tick_type_key_enum.TICKCOLOR: tick_color_enum.GREEN},
            order_forward_enum.MIDDLE: {tick_type_key_enum.TICKTYPE: tick_type_enum.OPENUNKOWN,
                                        tick_type_key_enum.TICKCOLOR: tick_color_enum.WHITE}
        },
    open_interest_delta_forward_enum.CLOSEFWDOUBLE:
        {
            order_forward_enum.UP: {tick_type_key_enum.TICKTYPE: tick_type_enum.CLOSEDOUBLE,
                                    tick_type_key_enum.TICKCOLOR: tick_color_enum.RED},
            order_forward_enum.DOWN: {tick_type_key_enum.TICKTYPE: tick_type_enum.CLOSEDOUBLE,
                                      tick_type_key_enum.TICKCOLOR: tick_color_enum.GREEN},
            order_forward_enum.MIDDLE: {tick_type_key_enum.TICKTYPE: tick_type_enum.CLOSEDOUBLE,
                                        tick_type_key_enum.TICKCOLOR: tick_color_enum.WHITE}
        },
    open_interest_delta_forward_enum.CLOSE:
        {
            order_forward_enum.UP: {tick_type_key_enum.TICKTYPE: tick_type_enum.CLOSESHORT,
                                    tick_type_key_enum.TICKCOLOR: tick_color_enum.RED},
            order_forward_enum.DOWN: {tick_type_key_enum.TICKTYPE: tick_type_enum.CLOSELONG,
                                      tick_type_key_enum.TICKCOLOR: tick_color_enum.GREEN},
            order_forward_enum.MIDDLE: {tick_type_key_enum.TICKTYPE: tick_type_enum.CLOSEUNKOWN,
                                        tick_type_key_enum.TICKCOLOR: tick_color_enum.WHITE}
        },
}

openclose_mapping = {
    '换手': open_interest_delta_forward_enum.EXCHANGE,
    '双开': open_interest_delta_forward_enum.OPENFWDOUBLE,
    '开仓': open_interest_delta_forward_enum.OPEN,
    '双平': open_interest_delta_forward_enum.CLOSEFWDOUBLE,
    '平仓': open_interest_delta_forward_enum.CLOSE,
    '无交易': open_interest_delta_forward_enum.NONE,
    '未知': open_interest_delta_forward_enum.UNKNOWN
}

price_direction_mapping = {
    '向上': order_forward_enum.UP,
    '向下': order_forward_enum.DOWN,
    '不变': order_forward_enum.MIDDLE
}

tick_type_to_chinese = {
    tick_type_enum.NOCHANGE: '无交易',
    tick_type_enum.UNKNOWN: '未知',
    tick_type_enum.EXCHANGELONG: '多换',
    tick_type_enum.EXCHANGESHORT: '空换',
    tick_type_enum.EXCHANGEUNKOWN: '未知换',
    tick_type_enum.OPENDOUBLE: '双开',
    tick_type_enum.OPENLONG: '开多',
    tick_type_enum.OPENSHORT: '开空',
    tick_type_enum.OPENUNKOWN: '未知开',
    tick_type_enum.CLOSEDOUBLE: '双平',
    tick_type_enum.CLOSESHORT: '空平',
    tick_type_enum.CLOSELONG: '多平',
    tick_type_enum.CLOSEUNKOWN: '未知平',
}

tick_color_to_chinese = {
    tick_color_enum.WHITE: '白色',
    tick_color_enum.RED: '红色',
    tick_color_enum.GREEN: '绿色',
}

# 加载数据
df = pd.read_csv(FILENAME+".csv")

# 增量计算
df['volume_delta'] = df['volume'].diff()
df['open_interest_delta'] = df['open_interest'].diff()
df['close_count'] = df['volume_delta']

# 前一笔的last_price
df['pre_last_price'] = df['last_price'].shift(1)

# 根据volume_delta和open_interest_delta确定openclose
# 使用np.select批量处理
conditions_openclose = [
    (df['open_interest_delta'] > 0) & (df['volume_delta'] == df['open_interest_delta']),
    (df['open_interest_delta'] > 0) & (df['volume_delta'] > df['open_interest_delta']),
    (df['open_interest_delta'] < 0) & (df['volume_delta'] > df['open_interest_delta'].abs()),
    (df['volume_delta'] > 0) & (df['open_interest_delta'] == 0),
    (df['volume_delta'] > 0) & ((df['volume_delta'] + df['open_interest_delta']) == 0),
    (df['volume_delta'] == 0) & (df['open_interest_delta'] == 0)
]
choices_openclose = ['双开','开仓','平仓','换手','双平','无交易']
df['openclose'] = np.select(conditions_openclose, choices_openclose, default='未知')

# 根据价格判断方向
# 若需要严格按照上一笔last_price来判断趋势
df['price_direction'] = np.where(df['last_price'] > df['pre_last_price'], '向上',
                                 np.where(df['last_price'] < df['pre_last_price'], '向下', '不变'))

# 去除第一行可能的NaN影响
df['openclose'] = df['openclose'].fillna('无交易')
df['price_direction'] = df['price_direction'].fillna('不变')

# 将openclose和price_direction映射到Enum
delta_enum_series = df['openclose'].map(openclose_mapping).fillna(open_interest_delta_forward_enum.NONE)
direction_enum_series = df['price_direction'].map(price_direction_mapping).fillna(order_forward_enum.MIDDLE)

# 基于delta_enum和direction_enum去确定tick_type和color
# 为了向量化，可以构建映射字典并使用apply/map
# 由于tick_type_cal_dict的嵌套结构复杂，可以先将其转为DataFrame或进行二维映射

# 首先，我们将tick_type_cal_dict转成方便查询的字典：
expanded_dict = {}
for delta_k, delta_v in tick_type_cal_dict.items():
    for dir_k, result_dict in delta_v.items():
        expanded_dict[(delta_k, dir_k)] = (result_dict[tick_type_key_enum.TICKTYPE], result_dict[tick_type_key_enum.TICKCOLOR])

# 定义一个取结果的函数
def get_tick_info(delta_enum, direction_enum):
    return expanded_dict.get((delta_enum, direction_enum), (tick_type_enum.UNKNOWN, tick_color_enum.WHITE))

# 使用apply对Series进行矢量化，注意Enum本身可能需要转成字符串key或直接传入
df['tick_type_color_tuple'] = list(map(get_tick_info, delta_enum_series, direction_enum_series))

# 拆分为两列
df['tick_type_enum'] = df['tick_type_color_tuple'].apply(lambda x: x[0])
df['tick_color_enum'] = df['tick_type_color_tuple'].apply(lambda x: x[1])

# 转为中文
df['trade_openclose'] = df['tick_type_enum'].map(lambda x: tick_type_to_chinese[x])
df['trade_color'] = df['tick_color_enum'].map(lambda x: tick_color_to_chinese[x])

# 清理多余列
df.drop(['tick_type_color_tuple','tick_type_enum','tick_color_enum','pre_last_price'], axis=1, inplace=True)

# 导出
df.to_csv(FILENAME+"_withopenclose.csv", index=False)
```

```python
import pandas as pd

FILENAME = "future_taobao_ssMain_tick"
# FILENAME = "future_taobao_ss2401_tick"

# 假设已加载并准备好DataFrame df
df = pd.read_csv(FILENAME+"_withopenclose.csv")

df = df.loc[df['volume_delta'] > 0]
# 重置索引以确保从0开始
df.reset_index(drop=True, inplace=True)


guess_tick=100
trade_gap = 5

df['dual_opportunity'] = 0
trade_hand = 1
trade_slippy = 0

# 为每个tick计算未来100个tick内与当前挂单价格相匹配的成交量
for i in range(len(df)):
    if i + guess_tick >= len(df):  # 确保未来有足够的tick进行分析
        break

    # 当前的买一和卖一价
    current_ask_price1 = df.loc[i, 'ask_price1']
    current_bid_price1 = df.loc[i, 'bid_price1']

    # 初始化对应价格的成交量计数
    matching_buy_volume = 0
    matching_sell_volume = 0
    
    buy_done = 0
    sell_done = 0
    stop = 0

    # 遍历未来100个tick
    for j in range(i+1, i+guess_tick+1):
        # if df.loc[j, 'bid_price1']<current_bid_price1-trade_gap or df.loc[j, 'ask_price1']>current_ask_price1+trade_gap:
        #     stop = 1
        #     break
        if buy_done ==1 and sell_done==1:
            break
        if df.loc[j, 'bid_price1']<current_bid_price1:
            buy_done = 1 
        if df.loc[j, 'ask_price1']>current_ask_price1:
            sell_done = 1 
        # 如果未来tick的买一价等于当前tick的买一价，累加成交量
        if df.loc[j, 'bid_price1'] == current_bid_price1 and df.loc[j, 'bid_price1'] == df.loc[j, 'last_price']:
            matching_buy_volume += df.loc[j, 'volume_delta']
            if matching_buy_volume >= df.loc[i, 'bid_volume1']+trade_hand+trade_slippy:
                buy_done = 1
        # 如果未来tick的卖一价等于当前tick的卖一价，累加成交量
        if df.loc[j, 'ask_price1'] == current_ask_price1 and df.loc[j, 'ask_price1'] == df.loc[j, 'last_price']:
            matching_sell_volume += df.loc[j, 'volume_delta']
            if matching_sell_volume >= df.loc[i, 'ask_volume1']+trade_hand+trade_slippy:
                sell_done = 1
    
    # if stop==1:
    #     df.loc[i,'dual_opportunity'] = 0
    if buy_done==1 and sell_done==1:
        df.loc[i,'dual_opportunity'] = 1
    else:
        df.loc[i,'dual_opportunity'] = 0


# 计算同时为买卖机会的总数
dual_opportunities_count = df['dual_opportunity'].sum()

# 计算比例
total_ticks_analyzed = len(df) - guess_tick
dual_opportunities_ratio = dual_opportunities_count / total_ticks_analyzed

print(f"Dual opportunities count: {dual_opportunities_count}")
print(f"Dual opportunities ratio: {dual_opportunities_ratio:.4f}")

# 保存到新的CSV文件
df.to_csv(FILENAME+"_with_opportunities.csv", index=False)

```

```python
import pandas as pd
import numpy as np  
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score,precision_score

FILENAME = "future_taobao_ssMain_tick"


# 加载数据
data = pd.read_csv(FILENAME+"_with_opportunities.csv")

data_clean = data.dropna().copy()  # 创建一个副本以避免警告

data_clean =  pd.get_dummies(data_clean, columns=['trade_color', 'trade_openclose'])

features = [
    'last_price','highest','lowest', 'volume', 'open_interest', 'volume_delta', 'open_interest_delta',
    'bid_price1', 'ask_price1', 'bid_price2', 'ask_price2', 'bid_price3', 'ask_price3',
    'bid_price4', 'ask_price4', 'bid_price5', 'ask_price5', 'bid_volume1', 'bid_volume2',
    'bid_volume3', 'bid_volume4', 'bid_volume5', 'ask_volume1', 'ask_volume2', 'ask_volume3',
    'ask_volume4', 'ask_volume5'
]
# 获取所有列名
all_columns = data_clean.columns.tolist()
# 筛选出以"trade_color_"和"trade_openclose_"开头的列名
# 并将它们添加到features列表中
for column in all_columns:
    if column.startswith('trade_color_') or column.startswith('trade_openclose_'):
        features.append(column)

window_size = 200  # 用于特征计算的窗口大小
# 特征生成
data_clean['rolling_mean'] = data_clean['last_price'].rolling(window=window_size).mean()
data_clean['rolling_std'] = data_clean['last_price'].rolling(window=window_size).std()
data_clean['ask_rolling_mean'] = data_clean['ask_price1'].rolling(window=window_size).mean()
data_clean['ask_rolling_std'] = data_clean['ask_price1'].rolling(window=window_size).std()
data_clean['bid_rolling_mean'] = data_clean['bid_price1'].rolling(window=window_size).mean()
data_clean['bid_rolling_std'] = data_clean['bid_price1'].rolling(window=window_size).std()
# RSI计算
delta = data_clean['last_price'].diff()
gain = (delta.where(delta > 0, 0)).fillna(0)
loss = (-delta.where(delta < 0, 0)).fillna(0)
avg_gain = gain.rolling(window=window_size).mean()
avg_loss = loss.rolling(window=window_size).mean()
rs = avg_gain / avg_loss
data_clean['RSI'] = 100 - (100 / (1 + rs))
# MACD计算
short_ema = data_clean['last_price'].ewm(span=int(window_size/2), adjust=False).mean()  # 使用窗口大小的一半作为短期EMA
long_ema = data_clean['last_price'].ewm(span=window_size, adjust=False).mean()  # 使用整个窗口大小作为长期EMA
data_clean['MACD'] = short_ema - long_ema
data_clean['MACD_signal'] = data_clean['MACD'].ewm(span=int(window_size/3), adjust=False).mean()  # 使用窗口大小的三分之一作为信号线
# 删除因计算滚动特征而产生的NaN值
data_clean.dropna(inplace=True)
# 添加新特征到特征列表
features.extend(['rolling_mean', 'rolling_std','ask_rolling_mean', 'ask_rolling_std','bid_rolling_mean', 'bid_rolling_std', 'RSI', 'MACD', 'MACD_signal'])

X = data_clean[['datetime'] + features]
y = data_clean['dual_opportunity']

# 数据切分
split_start = int(len(data_clean) * 0.3)
split_point = int(len(data_clean) * 0.8)
split_end = int(len(data_clean) * 1)
X_train = X.iloc[split_start:split_point]
y_train = y.iloc[split_start:split_point]
X_test = X.iloc[split_point:split_end]
y_test = y.iloc[split_point:split_end]
X_train_features = X_train.drop(columns=['datetime'])
X_test_features = X_test.drop(columns=['datetime'])



# 创建XGBoost模型并调整参数
model = xgb.XGBClassifier(
    n_estimators=50,  # 树的个数
    max_depth=3,  # 树的深度
    learning_rate=0.1,  # 学习率
    subsample=0.8,  # 训练每棵树时使用的样本比例
    colsample_bytree=0.8,  # 构建树时的列采样比例
    random_state=42,  # 随机种子
    use_label_encoder=False,  # 避免使用标签编码器的警告
    eval_metric='logloss',  # 评估指标
    scale_pos_weight=1
)
# 模型训练
model.fit(X_train_features, y_train)


# 获取属于类别1的概率
y_pred_proba = model.predict_proba(X_test_features)[:, 1]

# 设置交易阈值
trade_threshold = 0.65
# 根据阈值生成预测结果，大于等于阈值预测为1，否则为0
y_pred_threshold = (y_pred_proba >= trade_threshold).astype(int)
# 计算查准率
precision = precision_score(y_test, y_pred_threshold)
print(f"Precision: {precision}")
predicted_ones_count = np.sum(y_pred_threshold == 1)
total_count = len(y_pred_threshold)
print(f"Predicted as 1: {predicted_ones_count} / Total: {total_count}")

# 对照组评估（全1预测）
predictions_all_ones = [1] * len(y_test)
print("All Ones Accuracy:", precision_score(y_test, predictions_all_ones))
```

```python
from datetime import datetime, time
import numpy as np

guess_tick = 100
profit = 0  # 盈利
trade_threshold = 0.65 # 交易阈值
trade_hand = 1
trade_slippy = 0

trade_gap = 5 #每跳多少钱
hand_volume = 5  # 不锈钢1手5吨，1手开仓2元，平今0元
hand_fee = 2
done_count = 0
count = 0
daily_profit = {}  # 新增：用于存储每日盈利
daily_trade = {}

last_date_str=''
stop = False

# def check_trend_consistency(prices):
#     overall_trend_up = prices[-1] > prices[0]
#     for i in range(len(prices) - 1):
#         if overall_trend_up and prices[i + 1] < prices[i]:
#             return False
#         elif not overall_trend_up and prices[i + 1] > prices[i]:
#             return False
#     return True

def parse_time_range(time_range_str):
    """解析时间范围字符串并返回时间对象的开始和结束时间"""
    start_str, end_str = time_range_str.split('-')
    start_time = datetime.strptime(start_str, "%H:%M").time()
    end_time = datetime.strptime(end_str, "%H:%M").time()
    return start_time, end_time

def is_time_in_ranges(time_to_check, time_ranges):
    """判断给定时间是否在时间范围数组内"""
    for time_range in time_ranges:
        start_time, end_time = parse_time_range(time_range)
        if start_time <= time_to_check <= end_time:
            return True
    return False

# 定义时间范围数组
# notrade_time = ["08:30-09:20","10:00-10:40","11:20-11:30","13:30-13:40","14:50-15:00","21:00-21:10","0:00-1:00"]
notrade_time = ["08:30-09:10","10:10-10:30","11:25-11:30","14:55-15:00","20:30-21:10","0:50-1:00"]


i = guess_tick
while i < len(X_test) - guess_tick:  # 留出100个tick进行预测
    
    datetime_str = X_test.iloc[i]['datetime']
    date_str = datetime_str.split(' ')[0]  # 新增：提取日期字符串
    
    if last_date_str!= date_str: 
        last_date_str = date_str
        stop = False
    if stop:
        i = i + 1
        continue
    if daily_profit.get(date_str, 0)  < -100:
        last_date_str = date_str
        stop = True
    
    
    if '.' in datetime_str:
        # 如果有小数点，分割为主时间部分和纳秒部分
        time, nano_part = datetime_str.split('.')
    else:
        # 如果没有小数点，则没有纳秒部分
        time = datetime_str
        nano_part = '0'
    
    if is_time_in_ranges(datetime.strptime(time, "%Y-%m-%d %H:%M:%S").time(), notrade_time):
        i = i+1
        continue
    
    proba = y_pred_proba[i]
    if proba < trade_threshold:
        i = i+1
        continue
    
    # prices = X_test['ask_price1'].iloc[i-50:i].values
    # # 检查整体趋势一致性和变化范围
    # if (check_trend_consistency(prices) and abs(prices[-1] - prices[0])> 5)or abs(prices[-1] - prices[0])>10 :
    #     i += 100  # 注意：在for循环中直接修改i的值不会改变下一个迭代的索引
    #     continue
    
    
    current_ask_price1 = X_test.iloc[i]['ask_price1']
    current_bid_price1 = X_test.iloc[i]['bid_price1']
    
    if current_ask_price1-current_bid_price1!=trade_gap:
        i = i + 1
        continue

    matching_buy_volume = 0
    matching_sell_volume = 0
    buy_done = 0
    sell_done = 0
    
    buy_index = 0
    sell_index = 0
    buy_stop_index = 0
    sell_stop_index = 0
    
    for j in range(i + 1, i + guess_tick + 1): 
        future_tick = X_test.iloc[j]
        if future_tick['bid_price1'] < current_bid_price1:
            buy_done = 1
            if buy_index==0:
                buy_index = j
        if future_tick['ask_price1'] > current_ask_price1:
            sell_done = 1
            if sell_index==0:
                sell_index = j
        if future_tick['bid_price1'] == current_bid_price1 and future_tick['last_price'] == current_bid_price1:
            matching_buy_volume += future_tick['volume_delta']
            if matching_buy_volume >= X_test.iloc[i]['bid_volume1']+trade_hand+trade_slippy and buy_index==0:
                buy_index = j
        if future_tick['ask_price1'] == current_ask_price1 and future_tick['last_price'] == current_ask_price1:
            matching_sell_volume += future_tick['volume_delta']
            if matching_sell_volume >= X_test.iloc[i]['ask_volume1']+trade_hand+trade_slippy and sell_index==0:
                sell_index = j
        if buy_index>0 and sell_index>0:
            break
        if buy_index>0 and buy_stop_index==0 and future_tick['bid_price1']<current_bid_price1:
            buy_stop_index = j
            break
        if sell_index>0 and sell_stop_index==0 and future_tick['ask_price1']>current_ask_price1:
            sell_stop_index = j
            break

    stop_loss_tick = i + guess_tick
    stop_loss_price_bid = X_test.iloc[stop_loss_tick]['bid_price1']
    stop_loss_price_ask = X_test.iloc[stop_loss_tick]['ask_price1']
    count = count + 1
    daily_trade[date_str] = daily_trade.get(date_str,0)+1
    
    # 判断是否存在买卖机会
    if buy_index>0 and sell_index>0:
        trade_profit = (current_ask_price1 - current_bid_price1) * trade_hand * hand_volume-hand_fee*trade_hand
        profit += trade_profit
        daily_profit[date_str] = daily_profit.get(date_str, 0) + trade_profit  # 更新每日盈利
        done_count += 1
        print(X_test.iloc[i]['datetime'])
        print(trade_profit)
        print(X_test.iloc[buy_index]['datetime'])
        print(X_test.iloc[sell_index]['datetime'])
    elif buy_index>0:
        if buy_stop_index>0:
            trade_profit = (X_test.iloc[buy_stop_index]['bid_price1'] - current_bid_price1) * trade_hand * hand_volume-hand_fee*trade_hand
        else:
            trade_profit = (stop_loss_price_bid - current_bid_price1) * trade_hand * hand_volume-hand_fee*trade_hand 
        profit += trade_profit
        daily_profit[date_str] = daily_profit.get(date_str, 0) + trade_profit  # 更新每日盈利
        print(X_test.iloc[i]['datetime'])
        print(trade_profit)
        print(X_test.iloc[buy_index]['datetime'])
        if buy_stop_index>0:
            print(X_test.iloc[buy_stop_index]['datetime'])
        else:
            print(X_test.iloc[stop_loss_tick]['datetime'])
    elif sell_index>0:
        if sell_stop_index>0:
            trade_profit = (current_ask_price1 - X_test.iloc[sell_stop_index]['ask_price1'] ) * trade_hand * hand_volume-hand_fee*trade_hand
        else:
            trade_profit = (current_ask_price1 - stop_loss_price_ask) * trade_hand * hand_volume-hand_fee*trade_hand
        profit += trade_profit
        daily_profit[date_str] = daily_profit.get(date_str, 0) + trade_profit  # 更新每日盈利
        print(X_test.iloc[i]['datetime'])
        print(trade_profit)
        print(X_test.iloc[sell_index]['datetime'])
        if sell_stop_index>0:
            print(X_test.iloc[sell_stop_index]['datetime'])
        else:
            print(X_test.iloc[stop_loss_tick]['datetime'])
    i = stop_loss_tick
        

# 打印每一天的盈利
for date, daily_profit in daily_profit.items():
    print(f"{date}: {daily_profit:.2f}")
for date, daily_trade in daily_trade.items():
    print(f"{date}: {daily_trade}")

print(f"Final profit: {profit:.2f}")
print(done_count)
print(count)
print(done_count/count)

```
